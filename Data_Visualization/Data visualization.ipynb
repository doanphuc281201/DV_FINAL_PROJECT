### 2. Môn ngoại ngữ nào (ngoại trừ Tiếng Anh) có nhiều thí sinh tham gia nhất? Môn ngoại ngữ nào có điểm thi trung bình cao nhất?

```python
foreign_language=df[['Ma_mon_ngoai_ngu','Ngoai_ngu']][df['Ma_mon_ngoai_ngu']!='N1']

foreign_language=foreign_language.dropna()


fig = plt.figure(figsize =(10, 7))
plt.bar(['Tiếng Trung','Tiếng Nhật', 'Tiếng Pháp', 'Tiếng Nga', 'Tiếng Đức'], foreign_language['Ma_mon_ngoai_ngu'].value_counts(), color ='maroon')
 # x-axis : foreign_language['Ma_mon_ngoai_ngu'].value_counts().index
plt.xlabel("Mã môn ngoại ngữ")
plt.ylabel("Số thí sinh tham gia ")
plt.title("Số thí sinh dự thi các môn ngoại ngữ",fontweight ='bold')
plt.show()
```

```python
avarage_score=foreign_language['Ma_mon_ngoai_ngu'].value_counts().reset_index(name='count')
avarage_score['subject']=['Tiếng Trung','Tiếng Nhật', 'Tiếng Pháp', 'Tiếng Nga', 'Tiếng Đức']
avarage_score['mean']=np.nan
avarage_score
for i in range(len(avarage_score['subject'].unique())):
         avarage_score.loc[i,'mean']=np.mean(foreign_language[foreign_language['Ma_mon_ngoai_ngu']==avarage_score.loc[i,'index']]['Ngoai_ngu'])
avarage_score=avarage_score.sort_values ('mean', ascending=False)
fig = plt.figure(figsize =(10, 7))
plt.bar(avarage_score['subject'].tolist(), avarage_score['mean'].tolist())

plt.xlabel("Mã môn ngoại ngữ")
plt.ylabel("Điểm trung bình")
plt.title("Số thí sinh dự thi các môn ngoại ngữ",fontweight ='bold')
plt.show()
```

### 3. Tỉ lệ số lượng điểm 10 ở tất cả các tỉnh thành với cả nước và 10 tỉnh thành có số lượng điểm 10 cao nhất.

```python
df_10_points = df[(df["Toan"] == 10) | (df["Van"] == 10) | (df["Ngoai_ngu"] == 10)
             | (df["Li"] == 10) | (df["Hoa"] == 10) | (df["Sinh"] == 10)
             | (df["Su"] == 10) | (df["Dia"] == 10) | (df["GDCD"] == 10)].groupby(['Cum_thi']).size().reset_index(name='10_point')
df_10_points = df_10_points.sort_values(by=['10_point'], ascending=False)
fig = plt.figure(figsize=(20,15))
plt.pie(df_10_points.loc[: ,"10_point"], labels = df_10_points.loc[:, 'Cum_thi'], autopct="%1.2f%%")
plt.show();
```

```python
plt.figure(figsize=(12,8))
plt.bar(df_10_points.head(10)["Cum_thi"], df_10_points.head(10)["10_point"], align='center');
```

### 4. So sánh điểm trung bình tất cả các môn giữa TP. Hồ Chí Minh, TP. Hà Nội và cả nước

```python
df_HCM = df.loc[df["Cum_thi"] == "TP. Hồ Chí Minh"]
df_HN = df.loc[df["Cum_thi"] == "Hà Nội"]
columns = ["Dia", "GDCD", "Su", "Hoa", "Li", "Sinh", "Ngoai_ngu", "Toan", "Van"]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(df[columns].mean()))

plt.figure(figsize=(10, 10))
plt.subplot(polar=True)
plt.plot(label_loc, df[columns].mean(), label='Cả nước')
plt.plot(label_loc, df_HCM[columns].mean(), label='TP. Hồ Chí Minh')
plt.plot(label_loc, df_HN[columns].mean(), label='Hà Nội')

lines, labels = plt.thetagrids(np.degrees(label_loc), labels=columns)
plt.legend()
plt.show()
```

### 2. Phân tích hồi quy

```python
import statsmodels.api as sm
from statsmodels.api import Logit
```

```python
columns_Khoi_A = ["Toan", "Li", "Hoa"]
df_Khoi_A = df[columns_Khoi_A].dropna()
df_Khoi_A["Tong_diem"] = df_Khoi_A["Toan"] + df_Khoi_A["Li"] + df_Khoi_A["Hoa"]
df_Khoi_A["Ket_qua"] = np.where(df_Khoi_A["Tong_diem"] >= 27.2, 1, 0)
df_Khoi_A.head(5)
```

```python
x_Khoi_A = df_Khoi_A["Toan"]
y_Khoi_A =df_Khoi_A["Ket_qua"]
MLE_Khoi_A = Logit(y_Khoi_A, sm.add_constant(x_Khoi_A)).fit()
MLE_Khoi_A.summary()
```

```python
g = 1 / (1 + np.exp(48.1823 - 4.8775 * df_Khoi_A["Toan"].sort_values()))
plt.plot(df_Khoi_A["Toan"].sort_values(), g)
plt.xlabel("Điểm Toán")
plt.ylabel("Tỉ lệ đậu")
plt.title("Tỉ lệ đậu dựa vào điểm Toán xét theo khối A00")
plt.show()
```

```python
columns_Khoi_A1 = ["Toan", "Li", "Ngoai_ngu", "Ma_mon_ngoai_ngu"]
df_Khoi_A1 = df[columns_Khoi_A1].dropna()
df_Khoi_A1 = df_Khoi_A1.loc[df_Khoi_A1["Ma_mon_ngoai_ngu"] == "N1"]
df_Khoi_A1["Tong_diem"] = df_Khoi_A1["Toan"] + df_Khoi_A1["Li"] + df_Khoi_A1["Ngoai_ngu"]
df_Khoi_A1["Ket_qua"] = np.where(df_Khoi_A1["Tong_diem"] >= 27.2, 1, 0)
df_Khoi_A1.head(5)
```
